## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

airbnb.reviews_per_month.fillna(value=0, inplace= True)
airbnb['last_review'] = pd.to_datetime(airbnb['last_review'],format='%Y-%m-%d')
airbnb.last_review = airbnb['last_review'].dt.year.astype('object')
airbnb.last_review.fillna(value = 'never', inplace=True)
````

````python
airbnb.dropna(inplace=True)
airbnb.isnull().sum()
````

### What all manipulations have you done and insights you found?

#### Chart - 12 - Pair Plot

````python
# Pair Plot visualization code
plt.figure(figsize=(12,12))
pp = airbnb[['number_of_reviews','neighbourhood_group', 'room_type' , 'price','minimum_nights','availability_365','reviews_per_month']]
sns.pairplot(pp,hue='room_type');
````

The Seaborn Pairplot allows us to plot pairwise relationships between variables within a dataset. This creates a nice visualisation and helps us understand the data by summarising a large amount of data in a single figure.

The above charts shows that



* No_of_reviews vs price : we can see from above pairplot that as the price increases number of reviews decreases. and the lower price = higher number of reviews. highest number of reviews are getting by the listing . also highest
number of reviews is getting by the listing of private room type.

* Minimum_nights vs price: most host whose requirement of minimum night is high they prefer to put less price range.
